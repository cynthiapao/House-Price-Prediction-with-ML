```python
# import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
sns.set_context('notebook', font_scale=1.5)
%matplotlib inline

ames_train = pd.read_csv('Ames Housing dataset/train.csv')
ames_test = pd.read_csv('Ames Housing dataset/test.csv')

ames_train = pd.DataFrame(ames_train)
```

```python
ames_train.head(15)
```

```python
ames_train.describe()
```

```python
# deal with missing data
## drop columns where missing data occupy over 15%
missing = pd.DataFrame(ames_train.isnull().sum()/len(ames_train.index) * 100)
missing.columns = ['percent']
missing.loc[missing['percent'] > 15, 'column_select'] = True
missing.loc[missing['percent'] <= 15, 'column_select'] = False
missing.head()
```

```python
column_select = missing.index[missing['column_select'] == True].tolist()
column_select
```

```python
missing['column_select'].sum()
```

```python
ames_train.drop(column_select, inplace = True, axis = 1, errors = 'ignore')
ames_train.head(5)
```

```python
ames_train.isnull().sum()
```

How to deal with the left missing values?

- Let us try the simplest way first: drop them.

```python
from sklearn.preprocessing import LabelEncoder
ames_train_nona = ames_train.dropna()
ames_test_nona = ames_train.dropna()

def preprocessing(df, cols):
    # for categorical variables: relabel
    df_cat = df.select_dtypes(include=['object']).apply(LabelEncoder().fit_transform)
    # for continuous variables: normalization
    df_num = df._get_numeric_data().drop(['Id'], axis = 1)
    df_norm = (df_num - df_num.mean()) / (df_num.max() - df_num.min())
    df_new = pd.concat([df['Id'], df_norm, df_cat], axis=1)
    return df_norm, df_new
```

```python
ames_train_nona_norm, ames_train_nona = preprocessing(ames_train_nona, column_select)
ames_train_nona_norm, ames_test_nona = preprocessing(ames_test_nona, column_select)
ames_train_nona.head()
```

```python
# correlation analysis with SalePrice
corr = pd.DataFrame(ames_train_norm.corr(method = 'pearson')['SalePrice'])
corr.sort_values(['SalePrice'], ascending= False)
```

```python
cmap = sns.cubehelix_palette(light = 0.95, as_cmap = True)
plt.figure(figsize = (10, 10))
sns.heatmap(abs(ames_train_nona_norm.corr(method = 'pearson')), vmin = 0, vmax = 1, square = True, cmap = cmap)
```

```python
from xgboost import XGBClassifier
x_train = ames_train_nona.iloc[:, 1:-1].values
y_train = ames_train_nona['SalePrice'].values

x_test = ames_test_nona.iloc[:, 1:-1].values
y_test = ames_test_nona['SalePrice'].values

model = XGBClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(y_pred)
```
