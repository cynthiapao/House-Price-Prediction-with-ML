## Imputing Missing Data
- Drop columns where missing data occupy over 15%

```python
missing = pd.DataFrame(ames_train.isnull().sum()/len(ames_train.index) * 100)
missing.columns = ['percent']
missing.loc[missing['percent'] > 15, 'column_select'] = True
missing.loc[missing['percent'] <= 15, 'column_select'] = False
missing.head()
```

How to deal with the left missing values?

- Use mean values for numeric columns and the most frequent values for non-numeric columns

```python
from sklearn.base import TransformerMixin
class MissingDataImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        If the column is of dtype Object, then impute with the most frequent value.
        If the column is not of dtype Object, then impute with the median.

        """
    def fit(self, X, y=None):
        
        for col in X:
            if X[col].dtype == np.dtype('O'):
                self.fill = X[col].value_counts().index[0]
            else:
                self.fill = X[col].median()
                
        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
ames_traint = MissingDataImputer().fit_transform(ames_train)
ames_testt = MissingDataImputer().fit_transform(ames_test)

print('after...')
print(ames_testt.isnull().sum())
```
